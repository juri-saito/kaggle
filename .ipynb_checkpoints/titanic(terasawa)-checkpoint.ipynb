```python
#クラメール（質的*質的）
def CorrelationV(x, y):
    variation = ((y - y.mean()) ** 2).sum()
    inter_class = sum([((y[x == i] - y[x == i].mean()) ** 2).sum() for i in np.unique(x)])
    correlation_ratio = inter_class / variation
    return 1 - correlation_ratio
```

```python
CorrelationV(train['Survived'], train['Sex'])
```

```python
x = np.array([])
for i in ['Survived', 'Pclass', 'Sex']:
    for j in ['Survived', 'Pclass', 'Sex']:
        x = np.append(x,CorrelationV(train[i], train[j]))
print(pd.DataFrame(x.reshape(3,3)
                   ,columns = ['Survived', 'Pclass', 'Sex']
                   ,index = ['Survived', 'Pclass', 'Sex']))
```

* pclassとfareの相関係数はスピアマンの方が高い
* sibspとparchとfareの相関係数はスピアマンの方が上がる。
* 相関についてまとめられたものがあったので共有します
* ピアソンは正規分布を仮定（パラメトリック）で使用
* スピアマンは正規分布を仮定していない(ノンパラメトリック)

### ＊母平均の差の検定

```python
import scipy.stats as ss
```

```python
list_num = ['Sex', 'Age', 'SibSp', 'Parch', 'Fare']
for i in list_num:
    ttest = ss.ttest_ind(train[i].dropna(), test[i].dropna(), equal_var=False)
    print(f'----{i}:母平均の差の検定-----')
    print(f't値{ttest[0]}\np値{ttest[1]}')
```

### ＊独立性の検定

```python
# pclassの独立性の検定
train_test1 = train['Pclass'].value_counts().values
train_test1 = np.append(train_test1, test['Pclass'].value_counts().values)
train_test1 = train_test1.reshape(2,3)
chitest1 = ss.chi2_contingency(train_test1)
print('----Pclassのχ乗検定----')
print('χ二乗値', chitest1[0])
print('p値',chitest1[1])
print('自由度', chitest1[2])
print('期待度数表\n', chitest1[3])
```

```python
# Embarkedの独立性の検定
train_test2 = train['Embarked'].value_counts().values
train_test2 = np.append(train_test2, test['Embarked'].value_counts().values)
train_test2 = np.append(train_test2, 0)
train_test2 = train_test2.reshape(2,4)
chitest2 = ss.chi2_contingency(train_test2)
print('----Enbarkedのχ乗検定----')
print('χ二乗値', chitest2[0])
print('p値',chitest2[1])
print('自由度', chitest2[2])
print('期待度数表\n', chitest2[3])
```

* Embarkedのp値が低い　→ trainデータとtestデータで差が認められる

## Dataの準備

```python
label = train["Survived"]
# train.drop(["Survived"], axis=1, inplace=True)
train.drop(["Survived",'Embarked'], axis=1, inplace=True)
```

```python
print(train.shape, test.shape)
train.dtypes
```

```python
#category変数のdtypeをcategotyに変換
# cat_feat = ["Pclass", "Sex", "Embarked",'IsAlone']
cat_feat = ["Pclass", "Sex", 'IsAlone']
for i in cat_feat :
    train[i] = train[i].astype("category")
    test[i] = test[i].astype("category")
```

```python
train.head()
```

```python
train.shape
train.dtypes
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train, label)
```

```python
lgb_train = lgb.Dataset(X_train, y_train, categorical_feature=cat_feat)
lgb_eval = lgb.Dataset(X_valid, y_valid, reference=lgb_train, categorical_feature=cat_feat)

params = {
    "objective" : "binary",
    'learning_rate': 0.01,
    "metric" : "binary_logloss",
    "early_stopping_round" : 20
}
```

```python
model = lgb.train(params, lgb_train, valid_sets=[lgb_train, lgb_eval])
```

```python
val_pred = model.predict(X_valid, num_iteration=model.best_iteration)
```

```python
val_pred = np.where(val_pred < 0.5, 0, 1)
acc = accuracy_score(y_valid, val_pred)
acc
```

```python
cols = list(train.columns)
f_importance = model.feature_importance()
df_importance = pd.DataFrame({'feature':cols, 'importance':f_importance})
df_importance = df_importance.sort_values('importance', ascending=False) # 降順ソート
print('特徴量の重要度')
display(df_importance)
x = np.arange(1, len(df_importance['feature'].values)+1)
y = df_importance['importance'].values
plt.bar(x, y)
plt.xticks(x, df_importance['feature'].values.tolist())
plt.show()
```

```python
test = test.drop('Embarked',axis=1)
```

```python
test_pred = model.predict(test, num_iteration=model.best_iteration)
test_pred = np.where(test_pred < 0.5, 0, 1)
```
